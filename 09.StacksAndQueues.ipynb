## スタック (Stack)

- Pythonでは、`deque`を使うと効率的に実装できる

    - `pop`メソッド: 要素を取り出し
----
- Stack is a Last In First Out (LIFO) data structure
- In Python, `deque` can be used for efficient implementation
    - `append` method: add an element
    - `pop` method: remove and return an element

### Stackを使った文字列の反転: Reversing a String Using a Stack
- 文字列の各文字をスタックに追加し、スタックから取り出すことで反転した文字列を得る
- 先に入れた文字が後に出るため、文字列が逆順になる
----
- By pushing each character of a string onto a stack and then popping them off, we can obtain the reversed string
- The first character pushed will be the last one popped, resulting in the reversed order of the string


```python
def reverse_string(s: str) -> str:
    """
    文字列sをスタックを使って逆順にする関数
    Reverse the string s using a stack
    """
    stack: deque[str] = deque()# スタックの初期化: Initialize the stack
    # 文字列の各文字をスタックに追加
    # Push each character of the string onto the stack
    for char in s:
        stack.append(char)
    reversed_s:str = ''
    while stack:# stackが空でない間: while stack is not empty
        # スタックから文字を取り出して連結
        # Pop characters from the stack and concatenate
        reversed_s += stack.pop() 
    return reversed_s

input_string = "Hello, World!"
reversed_string = reverse_string(input_string)
print("Original String:", input_string)
print("Reversed String:", reversed_string)
```

### スタックを使った再帰的関数のシミュレーション: Simulating Recursive Functions Using a Stack
- 再帰的関数の呼び出しは、内部的にはスタックを使って管理されている
- 明示的にスタックを使うことで、再帰を避けることができる
----
- Recursive function calls are internally managed using a stack
- By explicitly using a stack, we can avoid recursion
----

#### 二項係数の計算: Calculating Binomial Coefficients
$$\begin{align*}
\binom{n}{k} &= \binom{n-1}{k-1} + \binom{n-1}{k} \quad (1 \leq k \leq n-1) \\
\binom{n}{0} &= 1 \quad (n \geq 0) \\
\binom{n}{n} &= 1 \quad (n \geq 0)
\end{align*}$$
- 二項係数は、再帰的な定義を持つ組み合わせの数を表す
- スタックに$n$と$k$をプッシュし、再帰的な呼び出しをシミュレートすることで、二項係数を計算できる
----
- The binomial coefficient is defined recursively as the number of combinations
- By pushing $n$ and $k$ onto a stack and simulating the recursive calls, we can calculate the binomial coefficient

```python
def binomial_stack(n:int, m:int)->int:
    """
    スタックを使って二項係数を計算する関数
    Calculate binomial coefficient using a stack
    """
    stack: deque[tuple[int, int]] = deque()
    stack.append((n, m))
    result:int = 0
    while stack:
        current_n, current_m = stack.pop()
        if current_m == 0 or current_m == current_n:
            result += 1
        else:
            stack.append((current_n - 1, current_m - 1))
            stack.append((current_n - 1, current_m))
    return result
```

## キュー (Queue)
- キューは、先入れ先出し (FIFO: First In First Out)データ構造
- 待ち行列とも言う
- Pythonでは、`deque`を使うと効率的に実装できる
    - `append`メソッド: 要素を追加
    - `popleft`メソッド: 先頭の要素を取り出し
----
- Queue is a First In First Out (FIFO) data structure
- Also known as a waiting line
- In Python, `deque` can be used for efficient implementation
    - `append` method: add an element
    - `popleft` method: remove and return the first element
----

### 簡単な例: A Simple Example
- 窓口の待ち行列をシミュレートする
- $p_\text{arrival}$: 到着確率
- $p_\text{service}$: サービス確率
- 待っている人の数を時間とともにプロットする
----
- Simulate a service counter queue
- $p_\text{arrival}$: arrival probability
- $p_\text{service}$: service probability
- Plot the number of people waiting over time

```python
n = 1000
p_arrival=.6
p_service=.7
queue: deque[int] = deque()
for k in range(n):
    if random.random() < p_arrival:
        queue.append(1)
    if random.random() < p_service and queue:
        queue.popleft()
    s = '*'*len(queue)
    print(f"{k:05d}: {s}")
```
