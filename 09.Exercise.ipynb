# 09 スタックとキュー: Stack and Queue

## 課題
数式を考え、左右のカッコが正しく対応しているかをスタックを用いて判定することを考える。
カッコの種類が複数ある場合に対応できるように実装しなさい。

Let's consider the problem of checking if the parentheses in a mathematical expression are properly matched. Implement a function that uses a stack to determine if the parentheses are correctly paired, even when there are multiple types of parentheses.

## 課題
Queueを使って、Merge Sortを実装しよう。データの長さは2のべき乗であると仮定する。
1. データを長さ1のリストに分割し、Queueに入れる。
2. Queueから2つのリストを取り出し、マージして、Queueに入れる。
3. Queue内のリストが1つになるまで、2を繰り返す。
これを実装し、整列が正しく行われていることを確認しなさい。

Let's implement Merge Sort using a Queue. We will assume that the length of the data is a power of 2. The steps are as follows:
1. Split the data into lists of length 1 and enqueue them.
2. Dequeue two lists, merge them, and enqueue the merged list.
3. Repeat step 2 until there is only one list left in the Queue. 
Implement this and verify that the sorting is done correctly.
